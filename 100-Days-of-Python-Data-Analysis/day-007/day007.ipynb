## Manually compute and visualize the coefficients for OLS, Ridge, Lasso, and Elastic Net regression on the given sample data

```python
# Re-importing necessary libraries after the reset
import numpy as np
import matplotlib.pyplot as plt

# Sample Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Calculate means
x_mean = np.mean(X)
y_mean = np.mean(y)

# OLS Calculation
numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean) ** 2)
beta_1_ols = numerator / denominator
beta_0_ols = y_mean - beta_1_ols * x_mean

# Ridge Regression (with lambda = 1)
lambda_ridge = 1
beta_1_ridge = numerator / (denominator + lambda_ridge)
beta_0_ridge = y_mean - beta_1_ridge * x_mean

# Lasso Regression (Approximate, with lambda = 0.1)
lambda_lasso = 0.1
beta_1_lasso = max(0, numerator - lambda_lasso) / denominator  # Simplified Lasso calculation
beta_0_lasso = y_mean - beta_1_lasso * x_mean

# Elastic Net Calculation (with alpha = 0.5, l1_ratio = 0.5)
alpha_elastic = 0.5  # Regularization strength
l1_ratio = 0.5  # Mix of L1 and L2 regularization
beta_1_elastic = (numerator - (l1_ratio * alpha_elastic)) / (denominator + (1 - l1_ratio) * alpha_elastic)
beta_0_elastic = y_mean - beta_1_elastic * x_mean

# Predictions
y_ols_pred = beta_0_ols + beta_1_ols * X
y_ridge_pred = beta_0_ridge + beta_1_ridge * X
y_lasso_pred = beta_0_lasso + beta_1_lasso * X
y_elastic_pred = beta_0_elastic + beta_1_elastic * X

# Plotting the results
plt.figure(figsize=(10, 6))

# Plot original data
plt.scatter(X, y, color='black', label='Data Points')

# Plot OLS Prediction
plt.plot(X, y_ols_pred, label='OLS (Manual)', linestyle='--')

# Plot Ridge Prediction
plt.plot(X, y_ridge_pred, label='Ridge (λ=1)', linestyle=':')

# Plot Lasso Prediction
plt.plot(X, y_lasso_pred, label='Lasso (λ=0.1)', linestyle='-.')

# Plot Elastic Net Prediction
plt.plot(X, y_elastic_pred, label='Elastic Net (α=0.5, L1 ratio=0.5)', linestyle='-')

# Formatting the plot
plt.xlabel('Feature (X)')
plt.ylabel('Target (y)')
plt.title('Manual OLS, Ridge, Lasso, and Elastic Net Regression')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
