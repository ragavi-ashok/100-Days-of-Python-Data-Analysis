##### Perform a mean-centered linear regression analysis to determine the relationship between years of experience and salary (in $1000s). Given a dataset of experience (in years) and corresponding salary values, implement a function to calculate the slope and intercept of the regression line. Then, plot the data points and the regression line. What is the equation of the line, and how does the salary change with each additional year of experience?"

##### Dataset: Years of Experience (x) and Salary (y) in $1000s
#####  x_values = [1, 2, 3, 4, 5, 6, 7]
#####  y_values = [30, 35, 40, 45, 50, 55, 60]

```python
import matplotlib.pyplot as plt

def mean_centered_linear_regression(x_values, y_values):
    n = len(x_values)
    
    # Calculate means of x and y
    mean_x = sum(x_values) / n
    mean_y = sum(y_values) / n
    
    # Calculate the slope (m) and intercept (b)
    numerator = sum((x - mean_x) * (y - mean_y) for x, y in zip(x_values, y_values))
    denominator = sum((x - mean_x) ** 2 for x in x_values)
    
    slope = numerator / denominator
    intercept = mean_y - slope * mean_x
    
    return slope, intercept

# Dataset: Years of Experience (x) and Salary (y) in $1000s
x_values = [1, 2, 3, 4, 5, 6, 7]
y_values = [30, 35, 40, 45, 50, 55, 60]

# Perform mean-centered linear regression
slope, intercept = mean_centered_linear_regression(x_values, y_values)

# Print the slope and intercept
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# Create points for the regression line
line_x = [min(x_values), max(x_values)]
line_y = [slope * x + intercept for x in line_x]

# Plot the data points
plt.scatter(x_values, y_values, color='blue', label='Data Points')

# Plot the regression line
plt.plot(line_x, line_y, color='red', label=f'Regression Line: y = {slope:.2f}x + {intercept:.2f}')

# Add labels and title
plt.xlabel('Years of Experience')
plt.ylabel('Salary (in $1000s)')
plt.title('Mean-Centered Linear Regression: Experience vs Salary')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
